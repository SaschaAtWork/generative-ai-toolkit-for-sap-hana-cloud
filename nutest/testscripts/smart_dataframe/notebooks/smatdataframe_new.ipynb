```python
from hana_ml import dataframe
from hana_ml.algorithms.pal.utility import DataSets
from gen_ai_hub.proxy.langchain import init_llm
llm = init_llm('gpt-4.1', temperature=0.0, max_tokens=512)
```

```python
address = "xxxxxxxx"
port = 443
user = "PALDEVUSER"
password = "xxxxx"
cc = dataframe.ConnectionContext(address=address, port=port, user=user, password=password)
hana_df, _, _, _ = DataSets.load_iris_data(cc)
hana_df = hana_df.to_head("ID")
```

```python
from hana_ai.smart_dataframe import SmartDataFrame

sdf = SmartDataFrame(hana_df)
```

```python
from hana_ai.tools.df_tools.fetch_tools import FetchDataTool
fetch_tool = FetchDataTool(connection_context=cc, return_direct=True)

sdf.configure(tools=[fetch_tool], llm=llm)
```

```python
sdf.ask("show me the first 5 rows of the dataframe")
```

```python
new_df = sdf.transform("show me the first 5 rows of the dataframe")
```
